```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single DataFrame
merge_df = pd.merge(study_results, mouse_metadata, on="Mouse ID", how="outer")

# Display the data table for preview
#display(mouse_metadata)
#display(study_results)
display(merge_df)
```

```python
# Checking the number of mice
mice_count = len(pd.unique(merge_df['Mouse ID']))
mice_count
```

```python
# Our data should be uniquely identified by Mouse ID and Timepoint
# Get the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
duplicate_mice = merge_df[merge_df.duplicated(subset=['Mouse ID', 'Timepoint'], keep=False)]['Mouse ID'].unique()
duplicate_mice
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
mice_duplicate = merge_df.loc[merge_df["Mouse ID"] == "g989"]
mice_duplicate
```

```python
to_remove = mice_duplicate.index
cleaned_data = merge_df.drop(to_remove)
cleaned_data
```

```python
# Checking the number of mice in the clean DataFrame.
clean_mice_count = len(pd.unique(cleaned_data['Mouse ID']))
clean_mice_count
```

## Summary Statistics

```python
tumor_regimen_mean = cleaned_data.groupby('Drug Regimen')['Tumor Volume (mm3)'].mean()
tumor_regimen_mean
```

```python
tumor_regimen_median = cleaned_data.groupby('Drug Regimen')['Tumor Volume (mm3)'].median()
tumor_regimen_median
```

```python
tumor_regimen_variance = cleaned_data.groupby('Drug Regimen')['Tumor Volume (mm3)'].var()
tumor_regimen_variance
```

```python
tumor_regimen_std = cleaned_data.groupby('Drug Regimen')['Tumor Volume (mm3)'].std()
tumor_regimen_std
```

```python
tumor_regimen_sem = cleaned_data.groupby('Drug Regimen')['Tumor Volume (mm3)'].sem()
tumor_regimen_sem
```

```python
type(tumor_regimen_sem)
```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
# Assemble the resulting series into a single summary DataFrame.
summary_stats_df = pd.DataFrame({"Mean Tumor Volume" : tumor_regimen_mean,
                                "Median Tumor Volume" : tumor_regimen_median,
                                "Tumor Volume Variance" : tumor_regimen_variance,
                                "Tumor Volume Std. Dev." : tumor_regimen_std,
                                "Tumor Volume Std. Err." : tumor_regimen_sem})
display(summary_stats_df)
```

```python
# A more advanced method to generate a summary statistics table of mean, median, variance, standard deviation,
# and SEM of the tumor volume for each regimen (only one method is required in the solution)

# Using the aggregation method, produce the same summary statistics in a single line
agg_summary = cleaned_data.groupby('Drug Regimen').agg({'Tumor Volume (mm3)': 
                                                         ['mean', 'median', 'var', 'std', 'sem']})
agg_summary
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin

# Start by getting the last (greatest) timepoint for each mouse
treatment_regimens = ["Capomulin", "Ramicane", "Infubinol", "Ceftamin"]
#filter_data = cleaned_data[cleaned_data["Drug Regimen"].isin(treatment_regimens)]
mouse_groupby_max = cleaned_data.groupby("Mouse ID")["Timepoint"].max()
mouse_groupby_df = mouse_groupby_max.reset_index()
display(mouse_groupby_df)

# Merge this group df with the original DataFrame to get the tumor volume at the last timepoint
last_tumor_df = mouse_groupby_df.merge(cleaned_data, how='left', on = ["Mouse ID", "Timepoint"])
display(last_tumor_df)
```

```python

# Put treatments into a list for for loop (and later for plot labels)
treatments = ['Capomulin', 'Ramicane', 'Infubinol', 'Ceftamin']

# Create empty list to fill with tumor vol data (for plotting)
tumor_vol = []

# Calculate the IQR and quantitatively determine if there are any potential outliers. 

    
    # Locate the rows which contain mice on each drug and get the tumor volumes
for regimen in treatments:
    regimen_data = last_tumor_df[last_tumor_df['Drug Regimen'] == regimen]['Tumor Volume (mm3)']
    tumor_vol.append(regimen_data)
    
    # add subset 
    quartiles = regimen_data.quantile([0.25, 0.5, 0.75])
    lower_q = quartiles[0.25]
    upper_q = quartiles[0.75]
    iqr = upper_q - lower_q
        
        # Determine outliers using upper and lower bounds
       
    lower_bound = lower_q - 1.5 * iqr
    upper_bound = upper_q + 1.5 * iqr
       
    outliers = regimen_data[(regimen_data < lower_bound) | (regimen_data > upper_bound)]
        
       
    print(f"{regimen}'s potential outliers: {outliers}")
```

```python
# Generate a box plot that shows the distrubution of the tumor volume for each treatment group.
plt.boxplot(tumor_vol, labels=treatments, flierprops = {'marker' :"o", 'markersize' : 12, 'markeredgecolor' : 'black', 'markerfacecolor' : 'r'})
      
plt.ylabel("Final Tumor Volume (mm3)")
plt.show
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and a linear regression model 
# for mouse weight and average observed tumor volume for the entire Capomulin regimen

capomulin_data = cleaned_data[cleaned_data["Drug Regimen"] == "Capomulin"]
average_tumor_volume = capomulin_data.groupby("Mouse ID")["Tumor Volume (mm3)"].mean()
capomulin_merged = pd.merge(capomulin_data, average_tumor_volume, on=["Mouse ID"])
capomulin_merged.rename(columns={"Tumor Volume (mm3)_y" : "Tumor Volume (mm3) Mean"})
x_weight = capomulin_data["Weight (g)"]
y_avg_tumor = capomulin_merged["Tumor Volume (mm3)_y"]

slope, int, rvalue, pvalue, std_error = st.linregress(x_weight, y_avg_tumor)
reg_data = slope * x_weight + int
plt.plot(x_weight, reg_data, color = 'r')

plt.scatter(x_weight, y_avg_tumor, marker = 'o', facecolors = 'royalblue', alpha = .8)
plt.xlabel("Weight (g)")
plt.ylabel("Average Tumor Volume (mm3)")
plt.show()
```
